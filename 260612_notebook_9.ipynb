## ベイズの定理

ベイズの定理は、結果である事象Bが観測されたときに原因となる事象Aの発生する条件付き確率を推定するために用いられる。

$$
\begin{align*}
  P(A \mid B) &= \frac{P(B \mid A) P(A)}{P(B)} \\
  P(B) &= P(B \mid A) P(A) + P(B \mid \bar{A}) P(\bar{A})
\end{align*}
$$

ある感染症への感染/非感染を原因となる事象A、検査での陽性/陰性を結果となる事象Bとして、2つのケースで陽性者の感染率$P(A|B)$を計算する。

- $A$が感染、$\bar{A}$が非感染を表す。$B$が陽性、$\bar{B}$が陰性を表す。
- 全人口の感染率を$P(A)$とする。
- 感染者が検査をしたときに陽性になる確率を$P(B|A)$、非感染者が検査をしたときに陽性になる確率を$P(B|\bar{A})$とする。
- 陽性率$P(B)$を計算する。
- 陽性者の感染率$P(A|B)$を計算する。

```python
# サイコロの目を1度振って出た目を確率変数として定義する
# Define the roll of the dice as a random variable.
prob = np.array([1 / 6, 1 / 6, 1 / 6, 1 / 6, 1 / 6, 1 / 6])
dice = np.array([1, 2, 3, 4, 5, 6])
cum_prob = np.cumsum(prob)
df = pd.DataFrame()
df["出目"] = dice
df["確率"] = prob
df["累積確率"] = cum_prob
print(df)
```

```python
# 確率関数を図示する。
# Illustrate the probability function.
plt.plot(dice, prob, "o", color="b")
plt.vlines(dice, 0, prob, colors="b")
plt.ylabel("確率")
plt.xlabel("出目")
plt.ylim(0, )
```

```python
# 累積分布関数を図示する
# # Illustrate the cumulative distribution function
plt.plot(dice, cum_prob, "o", color="b", markeredgecolor="b")
plt.plot(dice + 1, cum_prob, "o", color="none", markeredgecolor="b")
plt.hlines(cum_prob, dice, dice + 1, "b")
plt.xlim(1, 6.5)
plt.ylim(0, 1.1)
plt.ylabel("累積確率")
plt.xlabel("出目")
```

```python
# 期待値を計算する
# Calculate expected value.
E_x = sum(prob * dice)
print(f"確率変数の期待値：{E_x}")
```

```python
# 分散を計算する
# Calculate variance.
V_x = sum((dice - E_x) ** 2 * prob)
print(f"確率変数の分散：{V_x}")
```
